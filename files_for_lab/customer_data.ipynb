#### 5.  - What to do with the variables policy and policy_type?

If we assume that for policy the L always stand for the same value than we can get rid of the classification into personal, corporate & special as they are classified in the policy_type column.

```python
for col in categorical_df:
    print(categorical_df[col].value_counts())
```

```python
# Then this would be our cleaned column
categorical_df['policy']=categorical_df['policy'].str.replace(".*(?=L)","")
categorical_df['policy']
```

#### 6. Reduce categorical columns, merge output

```python
def cleanemploymentstatus(x):
    if 'Employed' in x:
        return "Employed"
    elif 'Unemployed' in x:
        return "Unemployed"
    else:
        return 'Other'
    
categorical_df['employmentstatus'] = categorical_df['employmentstatus'].apply(cleanemploymentstatus)
```

```python
def clean_vehicleclass(x):
    if 'Four-Door Car' in x:
        return "Four-Door Car"
    elif 'Two-Door Car' in x:
        return "Two-Door Car"
    elif 'SUV' in x:
        return "SUV"
    else:
        return 'Premium'
    
categorical_df['vehicle_class'] = categorical_df['vehicle_class'].apply(clean_vehicleclass)
```

```python
def clean_maritalstatus(x):
    if 'Married' in x:
        return "Married"
    else:
        return 'Not Married'
    
categorical_df['marital_status'] = categorical_df['marital_status'].apply(clean_maritalstatus)
```

```python
def clean_education(x):
    if 'High School or Below' in x:
        return "High School or Below"
    elif 'College' in x:
        return "College"
    else:
        return 'Academic'
    
categorical_df['education'] = categorical_df['education'].apply(clean_education)
```

```python
categoricals_train= X_train.select_dtypes(object)
categoricals_test= X_test.select_dtypes(object)
categoricals_train.head()

onehot_train = categoricals_train.drop(['response', 'coverage', 'employmentstatus','location_code', 'vehicle_size'], axis = 1)
onehot_test = categoricals_test.drop(['response', 'coverage', 'employmentstatus','location_code', 'vehicle_size'], axis = 1)
onehot_train.head()
```

##### 5.1 Ordinal Encoding for:
- coverage
- employmentstatus
- location_code
- vehicle size

```python
ordinal_train = categoricals_train.drop(['response', 'state', 'gender','education', 'marital_status', 'policy_type', 'policy', 'renew_offer_type', 'sales_channel', 'vehicle_class'], axis = 1)
ordinal_test = categoricals_test.drop(['response', 'state', 'gender','education', 'marital_status', 'policy_type', 'policy', 'renew_offer_type', 'sales_channel', 'vehicle_class'], axis = 1)
ordinal_train.head()
```

```python
for col in ordinal_train:
    print(ordinal_train[col].value_counts())
```

```python
def coverage(x):
    if 'Basic' in x:
        return "0"
    elif 'Extended' in x:
        return "1"
    else:
        return '2'
    
ordinal_train['coverage'] = ordinal_train['coverage'].apply(coverage)
ordinal_test['coverage'] = ordinal_test['coverage'].apply(coverage)
```

```python
def employmentstatus(x):
    if 'Employed' in x:
        return "0"
    elif 'Unemployed' in x:
        return "1"
    else:
        return '2'
    
ordinal_train['employmentstatus'] = ordinal_train['employmentstatus'].apply(employmentstatus)
ordinal_test['employmentstatus'] = ordinal_test['employmentstatus'].apply(employmentstatus)
```

```python
def location_code(x):
    if 'Suburban' in x:
        return "0"
    elif 'Rural' in x:
        return "1"
    else:
        return '2'
    
ordinal_train['location_code'] = ordinal_train['location_code'].apply(location_code)
ordinal_test['location_code'] = ordinal_test['location_code'].apply(location_code)
```

```python
def vehicle_size(x):
    if 'Medsize' in x:
        return "0"
    elif 'Small' in x:
        return "1"
    else:
        return '2'
    
ordinal_train['vehicle_size'] = ordinal_train['vehicle_size'].apply(vehicle_size)
ordinal_test['vehicle_size'] = ordinal_test['vehicle_size'].apply(vehicle_size)
```
